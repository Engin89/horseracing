```python
import numpy as np

def read_table(path, columns, nrows=100, **kwargs):
    df = pd.read_csv(path, header=None, sep='\t', nrows=nrows, encoding='latin-1', **kwargs)
    df = df.replace('\\N', np.nan)
    df.columns = columns
    return df
```
