```python
%matplotlib inline
import pandas as pd
from IPython.display import display
pd.options.display.max_columns = None
```

```python
horse_df = pd.read_csv("../data/eng/horse.csv")
horse_df["activated"] = pd.to_datetime(horse_df["activated"])
horse_df["yearsactive"] = horse_df["activated"].dt.year - horse_df["fodd"]
print(horse_df.shape)
horse_df.sample(n = 10)
```

```python
horse_df.boxplot(column = ["earnings"], by = ["yearsactive"], figsize = (10,7), sym = '')
```

```python
horse_df.hist(bins=100, figsize = (20,15))
```

```python
horse_df.nlargest(15, "fodd") # 14 horses are born in the future
```

```python
horse_df.describe()
```
